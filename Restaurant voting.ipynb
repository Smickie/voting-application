```python
import random
from IPython.display import clear_output


def get_names(): # Function to collect names from the class, making this useful in future bootcamps
    names = []
    while True:
        name = input("Enter a name (or type 'done' to finish): ")
        if name.lower() == 'done': # Breaks the loop, must be written in this way
            break
        names.append(name)
    return names

def get_restaurants(): # Function to collect restaurants from the class
    clear_output() # Clears output but stacks the inputs made by this function, useful when keeping track of the proposed restaurant
    restaurants = []
    while True:
        restaurant = input("Enter a restaurant name (or type 'done' to finish): ")
        if restaurant.lower() == 'done': # Breaks the loop, must be written in this way
            break
        restaurants.append(restaurant)
    return restaurants

def voting_process(names, restaurant_votes): # Function for the voting process itself
    num_rounds = 2 # Two rounds of voting, configurable
    clear_output()
    print("-- THE VOTE --")
    for restaurant, votes in restaurant_votes.items(): # Will display all the choices during voting
        print(f"{restaurant}: {votes} votes") 

    random.shuffle(names) # Shuffles the list of names, only does this once meaning two rounds in the same order

    for _ in range(num_rounds): # calls num_rounds
        for name in names:
            while True:
                print(f"\n{name}, please vote for a restaurant:")
                vote = input().strip().lower()
                matched = False
                for restaurant in restaurant_votes:
                    if restaurant.lower().startswith(vote):
                        restaurant_votes[restaurant] += 1 # Adds one to the voting dictionary in main() function below
                        matched = True
                        break
                if matched: # makes sure that the player votes for a valid resturant and prompts again if that's the case
                    break # returns to the for name in names loop
                else: 
                    print("Invalid vote. Please vote for a valid restaurant.\n")
            clear_output() # Clears the output to "update" the tally with every vote
            print("-- THE VOTE --")
            for restaurant, votes in restaurant_votes.items():
                print(f"{restaurant}: {votes} votes")
    
    return restaurant_votes


def main():
    
    names = get_names() # Create list of names
    
    restaurants = get_restaurants() # Create list of restaurants
    
    restaurant_votes = {restaurant: 0 for restaurant in restaurants} # Create dictionary with each resturant and vote
    
    final_votes = voting_process(names, restaurant_votes) # Initiate the voting process
    
    clear_output() # Code to print final results
    print("-- FINAL RESULTS --")
    for restaurant, votes in final_votes.items():
        print(f"{restaurant}: {votes} votes")

    sorted_votes = sorted(final_votes.items(), key=lambda x: x[1], reverse=True)
    
    highest_vote_count = sorted_votes[0][1]
    tied_restaurants = [restaurant for restaurant, votes in sorted_votes if votes == highest_vote_count]

    if len(tied_restaurants) > 1:
        print("\nTIE DETECTED: Rerun program between", "and/or ".join(tied_restaurants)) # no tie breaker vote yet, program instead prompts the user to rerun the program
    else:
        winner = tied_restaurants[0]
        print("\nWINNER:", winner)

if __name__ == "__main__":
    main()
```
